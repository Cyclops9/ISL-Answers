6. We will now explore (6.12) and (6.13) further.
(a) Consider (6.12) with p = 1. For some choice of y1 and λ > 0,
plot (6.12) as a function of β1. Your plot should confirm that
(6.12) is solved by (6.14).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ridge_opt(y,B,alpha):
    return (y - B)**2 + alpha * B**2
```

```python
B_values = ridge_opt(1,np.arange(-20,20),5)
Br = 2 / (1+10)
plt.scatter(x=np.arange(-20,20),y=B_values)
plt.axvline(x=Br,color='r')
```

(b) Consider (6.13) with p = 1. For some choice of y1 and λ > 0,
plot (6.13) as a function of β1. Your plot should confirm that
(6.13) is solved by (6.15)

```python
def lasso_opt(y,B,alpha):
    return np.power(y-B,2) + alpha * abs(B)

y = 1
alpha = 5
B = np.arange(-20,20)
results = lasso_opt(y,B,alpha)
Bl = 0
if y > alpha / 2:
    Bl = y-alpha/2
elif y < - alpha / 2:
    Bl = y+alpha/2
else:
    Bl = 0
plt.scatter(x=B,y=results)
plt.axvline(x=Bl)
```
